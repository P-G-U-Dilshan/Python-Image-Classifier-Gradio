```python
# 1. අවශ්‍ය දේවල් install කිරීම
# `transformers` කියන්නේ Hugging Face එකෙන් pre-trained models පාවිච්චි කරන්න දෙන library එක.
# `torch` කියන්නේ deep learning framework එක.
# `Pillow` කියන්නේ image processing වලට.
!pip install gradio transformers torch pillow -q

# 2. අවශ්‍ය libraries import කිරීම
import gradio as gr
from transformers import pipeline

print("Libraries සාර්ථකව import කළා!")

# 3. AI "විශේෂඥයා" (pipeline) load කිරීම
# අපි මෙතනදී Google එකේ "Vision Transformer" (ViT) කියන බලවත් image classification model එකක් ගන්නවා.
# මේ pipeline එකෙන් photo එකක් අරන්, ඒක model එකට දෙන්න ඕන විදියට හදලා, prediction එක අරන් දෙන ඔක්කොම වැඩ ටික කරලා දෙනවා.
classifier = pipeline("image-classification", model="google/vit-base-patch16-224")

print("AI Model (Vision Transformer) සාර්ථකව load කළා!")

# 4. Prediction කරන function එක ලිවීම
def classify_animal(image):
    # classifier එකට image එක දුන්නම, ඒක predictions list එකක් return කරනවා.
    predictions = classifier(image)

    # ප්‍රතිඵල ලස්සනට format කරලා user ට දෙන්න හදමු.
    # Gradio වල 'Label' component එකට මේ format එක ගැලපෙනවා.
    # ප්‍රතිඵලය dictionary එකක් විදියට return කරමු.
    # උදා: {'cat': 0.9, 'dog': 0.1}
    return {pred['label']: pred['score'] for pred in predictions}

# 5. Gradio Interface එක නිර්මාණය කිරීම
iface = gr.Interface(
    fn=classify_animal,
    # Input එක Image එකක් විදියට
    inputs=gr.Image(type="pil", label="Upload a photo"),
    # Output එක Label එකක් විදියට (මේක classification වලට නියමයි)
    outputs=gr.Label(num_top_classes=3, label="Predicted results"),
    title="AI recognition",
    description=" The model may not be 100% accurate.",

)

# 6. Web App එක Launch කිරීම
iface.launch(share=True, debug=True)
```
